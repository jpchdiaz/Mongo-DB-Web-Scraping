```python
import pandas
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://mars.nasa.gov/news/'
browser.visit(url)
```

```python
for x in range(1, 2):
    
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
#     titles = soup.find_all('h3')
    titles = soup.find_all('div', class_='list_text')
    
    for title in titles:
        print('------')
#         these work for me
        header = title.find('div', class_='content_title')
        headline = header.find('a').text
        texter = title.find('div', class_='article_teaser_body').text
        print(headline)
        print(texter)
# teacher example:
#         head3 = title.find('h3')
#         link = head3.find('a')
#         href = link['href']
#         book_title = link['title']
#         print(title)
#         print(href)
```
